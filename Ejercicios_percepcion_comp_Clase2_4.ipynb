```
import matplotlib.pyplot as plt
import numpy as np
```

```
#Cargamos imágenes infraexpuestas para aplicarles una transformación logarítmica. 
#Usando el parámetro alpha para cambiar la curva de la función logarítmica


x = plt.imread('underexposed1.jpg')
print(x.shape)
```

```
plt.imshow(x)
```

```
#La imagen tiene pixeles entre 0 y 255, los pasamos a entre 0 y 1 para facilitar la computación

x_1 = x.astype('double')/255.0
plt.imshow(x_1)
```

```
alpha = 1.5
max_r = np.max(x_1, axis = (0,1)) # calcula el máximo de cada canal RGB, axis=(0,1) para que calcule el máximo de cada matriz R G y B
max_s = np.array([1.0, 1.0, 1.0])
```

```
# fórmila de la transf. logarítmica
c = max_s / (np.log(1.0+(np.exp(alpha)-1)*max_r))
Y = c * np.log(1 + (np.exp(alpha)-1) * x_1)
```

```
plt.rcParams['figure.figsize'] = [10,15]
plt.subplot(2,2,1)
plt.axis('off')
plt.imshow(x);plt.title('Infraexpuesta')
plt.subplot(2,2,2)
plt.axis('off')
plt.imshow(Y);plt.title('transformación logarítmica alpha=%.2f' % alpha)
```

```
# Ahora aplicamos la trnasformacion exponencial a una imagen sobreeexpuesta
```

```
# la cargamos y normalizamos los pixeles a 0-1
sobrexpuesta = plt.imread('sobreexpuesta.jpg').astype('double')/255.0
#plt.imshow(sobrexpuesta)
alpha = 5000
max_r = np.max(sobrexpuesta, axis = (0,1))
max_s = np.array([1., 1., 1.])
c = max_s / ((1 + alpha) ** max_r -1)
salida = c * ((1 + alpha) ** sobrexpuesta - 1)
plt.rcParams['figure.figsize'] = [15, 15]
plt.subplot(2,2,3)
plt.axis('off')
plt.imshow(sobrexpuesta)
plt.title('Sobreexpuesta')
plt.subplot(2,2,4)
plt.axis('off')
plt.imshow(salida)
plt.title('Transformación Exponencial alpha=%.1f' % alpha)
```
